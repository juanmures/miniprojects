```python
import pandas as pd
from scipy.stats import linregress
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
# Read data from file
df = pd.read_csv("data/epa-sea-level.csv", date_parser="Year")
df.head()
```

```python
# Create scatter plot
plt.scatter(data=df, x="Year", y="CSIRO Adjusted Sea Level", color="blue")

# Create first line of best fit
line_a = linregress(df["Year"], df["CSIRO Adjusted Sea Level"])
year_all_data = range(df['Year'].min(), 2051, 1)
csiro_sea_level_all_data = line_a.intercept + line_a.slope * year_all_data

plt.plot(year_all_data, csiro_sea_level_all_data, color="red")
# Add labels and title
plt.title("Rise in Sea Level")
plt.xlabel("Year")
plt.ylabel("Sea Level (inches)")

# Create second line of best fit
df_recent_data = df.loc[df["Year"] >= 2000]
line_b = linregress(df_recent_data["Year"], df_recent_data["CSIRO Adjusted Sea Level"])
year_recent_data = range(df_recent_data['Year'].min(), 2051, 1)
csiro_sea_level_recent_data = line_b.intercept + line_b.slope * year_recent_data

plt.scatter(data=df_recent_data, x="Year", y="CSIRO Adjusted Sea Level", color="orange")
plt.plot(year_recent_data, csiro_sea_level_recent_data, color="yellow")
```
